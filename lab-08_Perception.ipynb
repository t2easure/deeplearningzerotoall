# perceptron
- 인공 신경망 : 인간의 뇌에서 신호를 전달하는 역할 수행하는 뉴런을 모방한 것 
- perceptron : 인공 신경망의 한 종류

입력 x가 들어왔을 때 x들의 가중치를 곱하게 되고, 그것들의 합과 bias를 더해서 output을 만듦.
- output은 activation function(시그모이드 등)에 들어감.
- 초창기 perceptron 은 linear classifier를 위해서 만들어짐.
(개와 고양이 분류 등 )

# AND, OR
- And : 둘다 1일 때만 1
- Or : 하나만 1이여도 1
- linear classifier로 분류 가능 : 그래프에서 직선 긋기 가능. 

# XOR
- 한개의 레이어를 가지는 perceptron은 xor를 풀 수 없고 여러 개의 레이어를 가져야 하는데, 각 층의 weight를 학습할 수 있는 방법이 없다는 것을 증명하게 됨. => 인공지능의 암흑기 (back propagation 나오기 전)
- 서로 다른 입력이 들어왔을 때 1, 아니면 0
- linear classifier로 불가능 : 하나의 직선을 그어서 결과의 0, 1 분류 불가능. 

아래는 xor 코드를 이용해 perceptron으로 xor 해결 가능한지를 알아봄.

```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

torch.manual_seed(777)
if device == 'cuda':
    torch.cuda.manual_seed_all(777)
```

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
```

```python
# nn layers
linear = torch.nn.Linear(2, 1, bias=True) # in, out
sigmoid = torch.nn.Sigmoid()

# model 
model = torch.nn.Sequential(linear, sigmoid).to(device)

# define cost/loss & optimizer
criterion = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)

for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)

    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()

    if step % 100 == 0:
        print(step, cost.item())
```

```python
# Accuracy computation

with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == Y).float().mean()
    print("\nHypothesis: ", hypothesis.detach().cpu().numpy(), "\nCorrect: ", predicted.detach().cpu().numpy())
```
